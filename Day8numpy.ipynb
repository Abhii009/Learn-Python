```python
#zeros
import numpy as np

#array 1d
arr = np.zeros(10)
print(arr)

#2d
arr1 = np.zeros((3,3)) #row and second colloums
print(arr1)

```

```python
#full
#arr 1d
arr = np.full(5,8)
print(arr)

#arr 2d
arr1 = np.full((3,3),5)
print( arr1)

#array 3d
arr2 = np.full((3,4,5),7)
print(arr2)

```

```python
#shape
arr = np.zeros(6) #6 zeros
print(arr)
print(arr.shape) #6
print(arr.ndim) #1d array
print(arr.size) #6
print(arr.dtype) #float

#2d
print("2d ones start")
arr1 = np.zeros((3,3))
print(arr1)
print(arr1.shape) #3,3
print(arr1.ndim) #2d array
print(arr1.size) #9
print(arr1.dtype) #float

```

```python
#reshape
#1d

arr = np.arange(6)

print(arr)

#2d
arr1 = arr.reshape(2,3)  #2 rows and 3 collums
print(arr1)

#3d
arr2 = arr.reshape(1,2,3) # Fixed: (1*2*3 = 6) to match the size of arr
print(arr2)
```

```python
#flatten
import numpy as np
#2d
arr = np.arange(9)
arr1 = arr.reshape(3,3)
print(arr1)

#flatten | 2d
arr2 = arr1.flatten()
print(arr2)
```

```python
#ravel
#2d
arr = np.arange(9).reshape(3,3)
print(arr)

#ravel
arr_new1 = arr1.flatten()
arr_new1[0] = 200
#print(arr.ravel())
print(arr_new1)
```

```python
#transpose  | isse arr change ho jata hai values ki line change ho jati hai
arr = np.arange(9).reshape(3,3)
print(arr)

#traspose
arr1 = arr.T
print(arr1)
```

```python
#stack | stacking
import numpy as np
#v-stack | vertical stack
arr_v1 = np.arange(9, 18) # Create another array for stacking
arr_v2 = np.arange(18, 27)
arr_v = np.vstack((arr_v1, arr_v2)) # Stack arr_v1 and arr_v2 vertically
print(arr_v)
'''vstack array ko row wise stack karta hai | vertical stack array ko ek ke neeche ek
jod deta hai eeth ki tarah'''
```

```python
#H-stack | Horizontally stack
arr_v1 = np.arange(9, 18) # Create another array for stacking
arr_v2 = np.arange(18, 27)
arr_v = np.hstack((arr_v1, arr_v2)) # Stack arr_v1 and arr_v2 vertically
print(arr_v)
'''Horizontal stack | hstack array ko collum wise stack karta hai | ek k aage ek jod deta hai'''
```

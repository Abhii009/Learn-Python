```python
#zeros
import numpy as np

#array 1d
arr = np.zeros(10)
print(arr)

#2d
arr1 = np.zeros((3,3)) #row and second colloums
print(arr1)

```

```python
#ones
#aray1d
arr = np.ones(5)
print(arr)

#array 2d
arr1 = np.ones((3,3))
print(arr1)
```

```python
#full
#arr 1d
arr = np.full(5,8)
print(arr)

#arr 2d
arr1 = np.full((3,3),5)
print( arr1)

#array 3d
arr2 = np.full((3,4,5),7)
print(arr2)

```

```python
#shape
arr = np.zeros(6) #6 zeros
print(arr)
print(arr.shape) #6
print(arr.ndim) #1d array
print(arr.size) #6
print(arr.dtype) #float

#2d
print("2d ones start")
arr1 = np.zeros((3,3))
print(arr1)
print(arr1.shape) #3,3
print(arr1.ndim) #2d array
print(arr1.size) #9
print(arr1.dtype) #float
```

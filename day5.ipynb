```python
#function calculator -> add , subtract
def cal(*a):
  print(eval(f"{a[0]} {a[-1]} {a[1]}"))
  #a[0] means a ka 0 index -> 10
  #a[-1] means a ka -1 (last) index
  #a[1] nmeans a ka 1 index -> 20

cal(10,20,"-")
cal(10,20,"+")
cal(10,20,"*")
cal(10,20,"/")
cal(10,20,"%")
cal(10,20,"//")

```

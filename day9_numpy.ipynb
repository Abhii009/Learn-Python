```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr[2])
print(arr[0:4])  #slicing
print(arr[-1])
print(arr[:3:2]) #start from 0 to 3 and jump 1
print(arr[::2])#jump by one
print(arr[::-1])#reverse
print(arr[3:5])
```

```python
#2D
arr = np.array([[1,2,3],[4,5,6]])
#print(arr)
# print(arr[0:2]) #row
# print(arr[:,0]) #all rows | collumns
print(arr[:,1]) #row > 1 | collumns | jitane bhi rows hai uska 1st element print karvaya hai hamne
print(arr[0,:])
```

```python
arr = np.array([[10,20],[30,40],])

print(arr[0,0])
print(arr[1,1])
print(arr[:,1])
print(arr[0,1,], arr[1,0]) #yaha [0,1] me 0th pura [10,20] bana hai, and 1th [30,40] | ab in dono ka 1st element print hua hai
print(arr[0,0,], arr[1,1])
```

```python
arr = np.array([[10,20],[30,40],[50,60]])
print(arr[0,1], arr[1,0], arr[2,0])
```

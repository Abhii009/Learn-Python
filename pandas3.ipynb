```python
import pandas as pd
import numpy as np
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan]
}
df = pd.DataFrame(data)
print(df)
# fill null values
df.isnull()


# sum of null values
df.isnull().sum()


# dropna
df.dropna()
print(df)

# fillna
df.fillna(5)
```

```python
# handle missing values
import pandas as pd
import numpy as np
data = {
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500]
}
df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
import numpy as np
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan]
}
df = pd.DataFrame(data)
print(df)

df.isnull()
```
